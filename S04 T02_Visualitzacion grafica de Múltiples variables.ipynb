```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
#Open choosen file

tips = pd.read_csv("Data-sources/tips.csv")
    
print(tips.head())
```

```python
tips.shape  # 244 rows, 7 columns
```

```python
tips.describe().round(2)  #estadísticos básicos con redondeo a 2 decimales
```

```python
# Create data
x= tips.total_bill  
y= tips.tip
colors="Blue"
area = np.pi*5
plt.axis([0,60,0,15])   #delimitamos el rango de cada eje

#plot
plt.scatter(x,y,s=area, c=colors, alpha=0.4, edgecolors="black")

plt.title("Total bill vs Tip")

plt.ylabel("Total bill")
plt.xlabel("Tip")
plt.show()
```

#### Cada uno de los puntos representa un caso. Hay un total de 244 observaciones

```python
# Matrix plot
sns.pairplot(tips, diag_kind="hist", plot_kws={"alpha": 0.2})

#La distribución de total bill es left-skeewed
```

```python
# Histogram

tips.hist(["tip"], bins=14)
plt.title("Total tips distribution")
plt.ylabel("Frequence")
plt.xlabel("Tips") 
```

```python
sns.set_style("darkgrid")
fig, ax = plt.subplots(figsize = (15, 5))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.title("Total bill distribution by day and by sex",fontsize=20,fontweight="bold")

# Viernes es el día de menos ocurrencias de total_bill, tambien por sexo.
```

```python
tips.day.unique()
# muestra los valores de la variable type en la variable day 
```

```python
sns.violinplot(data=tips,
x='size',
y='total_bill',
hue='sex',
split=True,   # will split different hues on different side of categories
bw=0.2,    # the smoothness of violin body
inner='quartile',   # show the quartile lines 
scale_hue=True,    # scale arcross each hue instead of all violins
scale='count') # scale the width of each violin
plt.title("Total bill distribution by sex",fontsize=16,fontweight="bold")
```

```python
fig, ax = plt.subplots(figsize = (15, 5))
sns.scatterplot(x = "total_bill", y = "tip", hue= "smoker", style= "smoker", data = tips
               )
plt.title("Bill distribution by smoker-status",fontsize=16,fontweight="bold")
```

```python
# Principales estadísticos
tips.total_bill.mean()
```

```python
tips.total_bill.mean() - tips.total_bill.std()
```

```python
tips.total_bill.std()
```

```python
# Gráfico con la distribución de cada una de las variables 
sns.jointplot("total_bill", "tip", data=tips, kind='reg');
```

## Nivell 2. Exercici 2
### Repeteix l'exercici 1 amb el dataset que disposem en el repositori de GitHub PRE-PROCESSING-DATA, movies.dat

```python
movies = pd.read_table('Pre-processing-data\movies.dat', sep="::", header = None)    # el archivo es .dat 
movies
```

```python
movies.columns
```

```python
movies.columns=["order","movie","genre"]  #ponemos nombres a las columnas
```

```python
movies.columns
```

```python
movies.genre.unique()
```

```python
# Separar movie de año
movies[["title","year"]] = movies["movie"].str.rsplit("(", n = 1, expand=True)
movies["year"] = movies["year"].str.replace(")","", regex=False)
movies["year"] = movies["year"].astype("int64")
movies
```

```python
movies.hist(["year"], bins=80)
plt.rcParams["figure.figsize"] = (15,5)
plt.title("Distribution of movies per year", fontsize=22, fontweight="bold")
plt.ylabel("Frequecy")
plt.xlabel("year")
```

```python
genre_dummies = movies["genre"].str.get_dummies(sep='|').add_prefix('col_')
genre_dummies
```

```python
genre_total = genre_dummies.sum(axis=0)
genre_total
```

```python
genre_total.plot(kind="bar", legend=False, figsize=(12, 4))
plt.title("Distribution of movies by genre", fontsize=22,fontweight="bold")
plt.xlabel("movie's genre")
plt.ylabel("movies(unit)")
```

```python
import squarify
sns.set(rc={"figure.figsize":(10,8)})
squarify.plot(sizes=genre_total[0:18], alpha=0.8, label=genre_total.index)
plt.axis('off')
plt.title("Distribution of movies by genre",fontsize=22
          ,fontweight="bold")
```

```python
movies.info()
```

```python
movies = pd.concat([movies,genre_dummies], axis = 1)   #concatenamos las variables dummies en el dataFrame movies
movies
```

```python
movies.info()
```

```python
movies_sum = movies.groupby("year")['col_Action', 'col_Adventure', 'col_Animation', "col_Children's",
       'col_Comedy', 'col_Crime', 'col_Documentary', 'col_Drama',
       'col_Fantasy', 'col_Film-Noir', 'col_Horror', 'col_Musical',
       'col_Mystery', 'col_Romance', 'col_Sci-Fi', 'col_Thriller', 'col_War',
       'col_Western'].sum()
plt.rcParams["figure.figsize"] = (15,5)

movies_sum.plot()
```

```python
#columna de total de películas por año
movies["total_g"]= movies.iloc[:,6:23].sum(axis=1)
movies.head()
```

```python
# column genre transformed into dummies variables
genre_dummies = movies["genre"].str.get_dummies(sep='|').add_prefix('col_')
genre_dummies
```

```python
sns.set_style("darkgrid")
plt.rcParams["figure.figsize"] = (10,5)
plt.title("Number of genres associated with each movie", fontsize=16, fontweight="bold")
ax = sns.countplot(x="total_g", data=movies)  #número de géneros asociados a cada película
```

```python
sns.displot(data=movies, x="year", kind="hist", multiple="fill", hue="total_g", height=8, aspect=1.0)
plt.title("Number of genre in movies per year", fontsize=16, fontweight="bold")
```

```python
movies_sum = movies.groupby('year', as_index=False)['col_Action',
       'col_Adventure', 'col_Animation', 'col_Children\'s', 'col_Comedy',
       'col_Crime', 'col_Documentary', 'col_Drama', 'col_Fantasy',
       'col_Film-Noir', 'col_Horror', 'col_Musical', 'col_Mystery',
       'col_Romance', 'col_Sci-Fi', 'col_Thriller', 'col_War', 'col_Western'].sum()
movies_sum.head()
```

```python
# nueva columna con el total de peliculas por año
movies_sum["total_year"] = movies_sum.iloc[:,1:19].sum(axis=1)
movies_sum.tail()
```

```python
movies_sum = movies.groupby("year")['col_Action', 'col_Adventure', 'col_Animation', "col_Children's",
       'col_Comedy', 'col_Crime', 'col_Documentary', 'col_Drama',
       'col_Fantasy', 'col_Film-Noir', 'col_Horror', 'col_Musical',
       'col_Mystery', 'col_Romance', 'col_Sci-Fi', 'col_Thriller', 'col_War',
       'col_Western']
movies_sum
```

```python
#agrupamos los años a décadas

movies["decade"] = (10 * (movies["year"] // 10)).astype(str) + 's'
movies.head()
```

```python
movies_decade = movies.groupby("decade")['col_Action', 'col_Adventure', 'col_Animation', "col_Children's",
       'col_Comedy', 'col_Crime', 'col_Documentary', 'col_Drama',
       'col_Fantasy', 'col_Film-Noir', 'col_Horror', 'col_Musical',
       'col_Mystery', 'col_Romance', 'col_Sci-Fi', 'col_Thriller', 'col_War',
       'col_Western'].sum()
movies_decade
```

```python
sns.set_theme()
f, ax = plt.subplots(figsize=(12, 6))
sns.heatmap(movies_decade, annot=True, fmt="d", linewidths=.5, ax=ax, cmap="icefire")
plt.title("Distribution of movies by decade and by genre", fontsize=16, fontweight="bold")
```
